```python
import re
def split_into_sentences(text):
    # Regular expression pattern, based on periods, question marks, and exclamation marks
    sentence_endings = r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s'
    return re.split(sentence_endings, text)    

from rouge_score import rouge_scorer

# Function to calculate ROUGE-L score using rouge-score library
def cal_rougeL(reference, hypothesis):
    scorer:rouge_scorer.RougeScorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    scores = scorer.score(reference, hypothesis)
    return scores['rougeL'].fmeasure  # Return only the f1-measure of ROUGE-L

def cal_rougeL_precision(reference, hypothesis):
    scorer:rouge_scorer.RougeScorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    scores = scorer.score(reference, hypothesis)
    return scores['rougeL'].precision  # Return only the f1-measure of ROUGE-L


def cal_rougeL_recall(reference, hypothesis):
    scorer:rouge_scorer.RougeScorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    scores = scorer.score(reference, hypothesis)
    return scores['rougeL'].recall  # Return only the f1-measure of ROUGE-L


for i in range(0, 4):    
    smollm_res = smollm_responses[i]
    remove_res = removellm_responses[i]
    ds = dataset[i]
    
    smollmRouge = {"f-score": dict(), "precision": dict(), "recall": dict()}
    removeRouge = {"f-score": dict(), "precision": dict(), "recall": dict()}
    
    for target in ['answer', 'mistake', 'hint']:
        smollmRouge['f-score'][target] = round(cal_rougeL(ds[target], smollm_res), 3)
        removeRouge['f-score'][target] = round(cal_rougeL(ds[target], remove_res), 3)
    
        smollmRouge['precision'][target] = round(cal_rougeL_precision(ds[target], smollm_res), 3)
        removeRouge['precision'][target] = round(cal_rougeL_precision(ds[target], remove_res), 3)
        
        smollmRouge['recall'][target] = round(cal_rougeL_recall(ds[target], smollm_res), 3)
        removeRouge['recall'][target] = round(cal_rougeL_recall(ds[target], remove_res), 3)
        
    if (i == 0):
        print("Normal Content")
    
    if (i == 1):
        print("Content + Mistake + Hint")
    
    if (i == 2):
        print("Content + Unrelevant")
    
    if (i == 3):
        print("Partial Content")
    
    print(f"smollm Fscore, {smollmRouge['f-score']}")
    print(f"smollm precis, {smollmRouge['precision']}")
    print(f"smollm recall, {smollmRouge['recall']}")
    print()
    print(f"remove Fscore, {removeRouge['f-score']}")
    print(f"remove precis, {removeRouge['precision']}")
    print(f"remove recall, {removeRouge['recall']}")
    print("____________________")
    print()
```
